__Feature Importance+ build model  + Score__

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

__split data__

```python
x = df.drop('price_range',axis = 1 )
y = df['price_range']
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 42)
```

__Train model__

```python
model = RandomForestClassifier()
model.fit(x_train,y_train)
```

__check accuracy__

```python
print('Train accuracy:',model.score(x_train,y_train))
print('Test accuracy:',model.score(x_test,y_test))
```

```python
# Feature importance plot 
feat_importance = pd.Series(model.feature_importances_, index = x.columns)
feat_importance.nlargest(10).plot(kind = 'barh',figsize=(10,6),color = 'green')
plt.title('Top 10 fetures importance')
plt.xlabel('Importance')
plt.ylabel('Features')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

__Build KNN__

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=10)
```

```python
knn.fit(x_train,y_train)
```

```python
print('Train accuracy:',knn.score(x_train,y_train))
print('Test accuracy:',knn.score(x_test,y_test))
```

__confusion metrics for knn__

```python
from sklearn.metrics import confusion_matrix 
y_pred = knn.predict(x_test)
cm_0 = confusion_matrix (y_test,y_pred)
cm_0
```

__Build SVM model__

```python
from sklearn.svm import SVC
```

```python
model = SVC( gamma=1,kernel = 'linear') 
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

__Confusion Metrics for SVM with heatmap__


```python
from sklearn.metrics import confusion_matrix
```

```python
y_pred = model.predict(x_test)
cm= confusion_matrix(y_test,y_pred)
plt.figure(figsize = (8,6))
sns.heatmap(cm,annot = True,cmap='Blues',fmt = 'g')
plt.xlabel('predicted')
plt.ylabel('actual')
plt.show()
```

```python
from sklearn.metrics import classification_report 
```

```python
y_pred = model.predict(x_test)
print(classification_report(y_test,y_pred))
```
